```python
X =  df.iloc[:,1:].values
Y = df.iloc[:,0].values
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

level_one_data = list()
for train_index, valid_index in kf.split(X_dev):
    array = list() 
    X_train, X_valid = X_dev[train_index], X_dev[valid_index]
    y_train, y_valid = Y_dev[train_index], Y_dev[valid_index]
    
    #Model 1
    reg = LinearRegression().fit(X_train, y_train.ravel())
    predicted_1 = reg.predict(X_valid)
    array.append(predicted_1)
    
    #Model 2
    rfg = RandomForestRegressor().fit(X_train, y_train.ravel())
    predicted_2 = rfg.predict(X_valid)
    array.append(predicted_2)
    
    #Model 3
    gb = GradientBoostingRegressor().fit(X_train, y_train.ravel())
    predicted_3 = gb.predict(X_valid)
    array.append(predicted_3)
    
    #True values
    array.append(y_valid)
    level_one_data.append(array)

level_one_data = np.array(level_one_data)
```
