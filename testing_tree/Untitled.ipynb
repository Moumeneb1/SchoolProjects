```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss='log', penalty='l1',max_iter=10000, l1_ratio=0.9, learning_rate='optimal', shuffle=True, n_jobs=3, fit_intercept=True)
clf.fit(X, y)
```
