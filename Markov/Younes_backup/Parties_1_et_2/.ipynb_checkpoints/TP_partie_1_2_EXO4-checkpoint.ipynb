#### Partie 4 : froid ou chaud ?

```python
# -*- coding: utf-8 -*-
"""

@author: barthes
"""
import matplotlib.pyplot as plt
import numpy as np
import scipy.io.matlab as mio
import pomegranate as pg

from duree import duree

    
    
gamma=0.95
alpha=0.65 
beta=0.02 
Nsamples = 100000
# Définition des paramétres du modéle
start_probability = np.array([1,0,0,0,0,0,0])


T = np.array([[0.6,0.4,0,0,0,0,0],
            [0.4/3,0.6,0.4/3,0.4/3,0,0,0],
            [0,0.2,0.6,0,0.2,0,0],
            [0,0.2,0,0.6,0.2,0,0],
            [0,0,0.4/3,0.4/3,0.6,0.4/3,0],
            [0,0,0,0,0.2,0.6,0.2],
            [0,0,0.,0,0,0.4,0.6]])               # Matrice de transition temporaire

B = np.array([[0.5,0.5,0,0,0,0,0],
            [0.5/3,0.5,0.5/3,0.5/3,0,0,0],
            [0,0.25,0.5,0,0.25,0,0],
            [0,0.25,0,0.5,0.25,0,0],
            [0,0,0.5/3,0.5/3,0.5,0.5/3,0],
            [0,0,0,0,0.25,0.5,0.25],
             [0,0,0.,0,0,0.5,0.5]])# matrice d'émission temporaire


dicoObs={'obsp1':0,'obsp2':1,'obsp3':2,'obsp4':3,'obsp5':4,'obsp6':5,'obsp7':6}        # pour transformer les chaines en entier (0 et 1)
dicoState= {'p1':0,'p2':1,'p3':2,'p4':3,'p5':4,'p6':5,'p7':6}

## Creation de la chaine de Markov
model = pg.HiddenMarkovModel( name="exercice04" )      # Creation instance
# Matrice d'emission

# Creation des etat prob emission
p1 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[0,0],'obsp2': B[0,1],'obsp3':B[0,2],'obsp4':B[0,3],'obsp5':B[0,4],'obsp6':B[0,5],'obsp7':B[0,6]}), name='p1' )
p2 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[1,0],'obsp2': B[1,1],'obsp3':B[1,2],'obsp4':B[1,3],'obsp5':B[1,4],'obsp6':B[1,5],'obsp7':B[1,6]}), name='p2' )
p3 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[2,0],'obsp2': B[2,1],'obsp3':B[2,2],'obsp4':B[2,3],'obsp5':B[2,4],'obsp6':B[2,5],'obsp7':B[2,6]}), name='p3' )
p4 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[3,0],'obsp2': B[3,1],'obsp3':B[3,2],'obsp4':B[3,3],'obsp5':B[3,4],'obsp6':B[3,5],'obsp7':B[3,6]}), name='p4' )
p5 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[4,0],'obsp2': B[4,1],'obsp3':B[4,2],'obsp4':B[4,3],'obsp5':B[4,4],'obsp6':B[4,5],'obsp7':B[4,6]}), name='p5' )
p6 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[5,0],'obsp2': B[5,1],'obsp3':B[5,2],'obsp4':B[5,3],'obsp5':B[5,4],'obsp6':B[5,5],'obsp7':B[5,6]}), name='p6' )
p7 = pg.State( pg.DiscreteDistribution({ 'obsp1': B[6,0],'obsp2': B[6,1],'obsp3':B[6,2],'obsp4':B[6,3],'obsp5':B[6,4],'obsp6':B[6,5],'obsp7':B[6,6]}), name='p7' )


           
           
           
           
           
# Matrice de transition
model.add_transitions(model.start,[p1,p2,p3,p4,p5,p6,p7],start_probability)  # Probs initiales 

model.add_transitions(p1, [p1,p2,p3,p4,p5,p6,p7], [T[0,0],T[0,1],T[0,2],T[0,3],T[0,4],T[0,5],T[0,6]])     # transitions depuis sunny
model.add_transitions(p2, [p1,p2,p3,p4,p5,p6,p7], [T[1,0],T[1,1],T[1,2],T[1,3],T[1,4],T[1,5],T[1,6]])     # transitions depuis sunny
model.add_transitions(p3, [p1,p2,p3,p4,p5,p6,p7], [T[2,0],T[2,1],T[2,2],T[2,3],T[2,4],T[2,5],T[2,6]])     # transitions depuis sunny
model.add_transitions(p4, [p1,p2,p3,p4,p5,p6,p7], [T[3,0],T[3,1],T[3,2],T[3,3],T[3,4],T[3,5],T[3,6]])     # transitions depuis sunny
model.add_transitions(p5, [p1,p2,p3,p4,p5,p6,p7], [T[4,0],T[4,1],T[4,2],T[4,3],T[4,4],T[4,5],T[4,6]])     # transitions depuis sunny
model.add_transitions(p6, [p1,p2,p3,p4,p5,p6,p7], [T[5,0],T[5,1],T[5,2],T[5,3],T[5,4],T[5,5],T[5,6]])     # transitions depuis sunny
model.add_transitions(p7, [p1,p2,p3,p4,p5,p6,p7], [T[6,0],T[6,1],T[6,2],T[6,3],T[6,4],T[6,5],T[6,6]])     # transitions depuis sunny

            

model.bake( verbose=True )      # A mettre lorsque la description de la chaine est finalisée



```
